## Extract text from PDF file and WordCloud
1. using textract

```python
path = "/Users/eunji/Desktop/rename/nonadia.docx"
```

## 1. using textract

```python
import textract
text = textract.process(path,language='eng',input_encoding=None).decode('utf-8')
print(text[:50])
print(type(text))
```

## Extracting Keywords

```python
import re
keywords = re.findall(r'[a-zA-Z]\w+',text)
keytostr = ' '.join(keywords)
print(len(keywords))
print(len(keytostr))
print(keytostr[:100])
print(type(keytostr))
```

2. WordCloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
#Generate a word cloud image
wordcloud = WordCloud(max_font_size=100).generate(text)
#Display
fig = plt.figure()
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
```

```python
import numpy as np
from PIL import Image
from wordcloud import STOPWORDS
bird_back = np.array(Image.open("bird.png"))
stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
wc = WordCloud(background_color="white",max_words=2000,mask=bird_back,stopwords=stopwords)
wc = wc.generate(text)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show()
```
