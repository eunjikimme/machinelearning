# html

```python
url = 'https://www.bbc.com/news/world-asia-52232147'
```

```python
import requests
# make the request and check object type
r = requests.get(url)
type(r)
# extract HTML from Response object and print
html = r.text
#print(html)
```

```python
# Import BeautifulSoup from bs4
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html5lib")
#type(soup)
#soup.title
soup.title.string
```

```python
naver_text = soup.get_text()
```

# Tokenize

re or nltk

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(naver_text)
tokens[:8]
```

```python
words = []
for word in tokens:
    words.append(word.lower())
words[:8]
```

# Remove stop words

```python
import nltk
sw = nltk.corpus.stopwords.words('english')
sw[:5]
```

```python
words_wo = []
for word in words:
    if word not in sw:
        words_wo.append(word)
words_wo[:5]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
#create freq dist and plot
freqdist1 = nltk.FreqDist(words_wo)
freqdist1.plot(25)
```
